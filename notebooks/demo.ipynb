### Initialize the model

```python
import torch
from src import CustomUNet
from torchviz import make_dot
from IPython.core.display import SVG

# Initialize the model
model = CustomUNet(input_shape=(1, 128, 256), num_classes=1)
model.to(torch.device("cuda"))
x = torch.randn(1, 1, 128, 256).to(torch.device("cuda"))
y = model(x)

dot = make_dot(y, params=dict(list(model.named_parameters())))
dot.format = 'svg'
graph = dot.render(filename='torchviz', directory='./', cleanup=True)
SVG(filename='torchviz.svg')
```

```python
# Set the random seed for reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
```
